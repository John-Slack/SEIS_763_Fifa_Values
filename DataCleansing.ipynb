```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

players = pd.read_csv("fifaValue.csv", delimiter=',', header=0)
players.head()
```

```python
players.describe(include='all')
```

```python
#Removing Special Character from Variable “Value”

import decimal
df_new = players['Value'].str.replace('[�]','')

# Converting K to Mill if Required
new_Value = []
for i in df_new:
  if 'K' in i:
    data = str(round(decimal.Decimal(int(i.replace('K',''))/1000000),6)) + 'M'
    new_Value.append(data)
  else:
    new_Value.append(i)

players["NRM_Value"] = new_Value
```

```python
# Convert Year and replaced Null with NA
from os import error
from datetime import datetime
import numpy as np

new_CVU = []

for each in players["Contract Valid Until"]:
  try:
    if len(each) > 4 :
      data = '20' + each[7:9]
      new_CVU.append(data)
    else:
      data = each
      new_CVU.append(each)
  except Exception :
    data = 'NA'
    new_CVU.append(data)
  
players["NRM_CVU"] = new_CVU

players.to_csv("NRM_Data.csv")
```
